# Ch1. Transformers, what can they do?
## Pipeline function

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "An event horizon is the theoretical, spherical boundary surrounding a black hole beyond which the gravitational pull is so intense that nothing, not even light, can escape",
    candidate_labels=["business", "astronomy", "education"],
)
```

```python
from transformers import pipeline
generator = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M")
thoughts = generator(
    "Today I should",
    max_length=30,
    num_return_sequences=1,
)
"""
[{'generated_text': 'In this course, we will teach you how to manipulate the world and '
'move your mental and physical capabilities to your advantage.'}]
"""
print(thoughts[0]['generated_text'])
```

```python
from transformers import pipeline
question_answerer = pipeline("question-answering")
result = question_answerer(
    question="Where does Sabi lives?",
    context="I have a new girlfriend, her name is Sabi and she lives Japan.",
)
#return a dictionary with keys
print(result["answer"])
```

```python
from transformers import pipeline
ner = pipeline("ner", aggregation_strategy="simple")
ner("Sabi is working at TypeMoon in Japan")
```

```python
from transformers import pipeline
image_classifier = pipeline(
    task="image-classification", model="google/vit-base-patch16-224"
)
result = image_classifier(
    "https://cdn.britannica.com/80/232780-050-404D6708/Pembroke-welsh-corgi-dog.jpg"
)
print(result)
```

```python
from transformers import pipeline
unmasker = pipeline("fill-mask", model="bert-base-uncased")
result = unmasker("This man works as a [MASK].")
print(list(r['token_str'] for r in result))
for r in result:
  print(r['sequence'])
```

# Ch2. Using Transformer


tokenizer -> model -> post-processing

inside a model:
- embedding Layer — “Turn tokens into vectors”
- Transformer Layers — “Contextualize everything”
- Hidden States — “The evolving token representation, understanding”
- Head — “Convert hidden states to task output”

```python
#Tokenizer
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

#Model
from transformers import AutoModel

model = AutoModel.from_pretrained(checkpoint)
```

```python
raw_inputs = [
    "Sabi is so sarcastic",
    "She is so pretty though",
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
print(inputs)
```

```python
outputs = model(**inputs)
print(outputs.last_hidden_state.shape)
```

```python
from transformers import AutoModelForSequenceClassification #head -> task-specific projection

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model2 = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model2(**inputs)
print(f"labels: {model2.config.id2label}")
print(f"output_example: {outputs}")
"""
logits shape is 2x2 since we have 2 sentences and two labels
"""

import torch #logits -> softmax function -> propability

predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(f"predictions: {predictions}")
```
