# Transformers, what can they do?
## Pipeline function

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "An event horizon is the theoretical, spherical boundary surrounding a black hole beyond which the gravitational pull is so intense that nothing, not even light, can escape",
    candidate_labels=["business", "astronomy", "education"],
)
```

```python
from transformers import pipeline
generator = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M")
thoughts = generator(
    "Today I should",
    max_length=30,
    num_return_sequences=1,
)
"""
[{'generated_text': 'In this course, we will teach you how to manipulate the world and '
'move your mental and physical capabilities to your advantage.'}]
"""
print(thoughts[0]['generated_text'])
```

```python
from transformers import pipeline
question_answerer = pipeline("question-answering")
result = question_answerer(
    question="Where does Sabi lives?",
    context="I have a new girlfriend, her name is Sabi and she lives Japan.",
)
#return a dictionary with keys
print(result["answer"])
```

```python
from transformers import pipeline
ner = pipeline("ner", aggregation_strategy="simple")
ner("Sabi is working at TypeMoon in Japan")
```

```python
from transformers import pipeline
image_classifier = pipeline(
    task="image-classification", model="google/vit-base-patch16-224"
)
result = image_classifier(
    "https://cdn.britannica.com/80/232780-050-404D6708/Pembroke-welsh-corgi-dog.jpg"
)
print(result)
```
