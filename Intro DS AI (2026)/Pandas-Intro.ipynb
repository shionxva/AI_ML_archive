### Import and create dataframe

```python
import pandas as pd

df = pd.DataFrame(
    {
        "Name": [
            "Braund, Mr. Owen Harris",
            "Allen, Mr. William Henry",
            "Bonnell, Miss. Elizabeth",
        ],
        "Age": [22, 35, 58],
        "Sex": ["male", "male", "female"],
    }
)

df
```

```python
# sort by name
df = df.sort_values(by ='Name')
```

```python
# take column Age
df['Age']

# round of the Age
df['Age rounded'] = (df['Age'] // 10) * 10

df
```

### Indexing (loc & iloc)

```python
import pandas as pd

data = {
    'A': [10, 20, 30, 40, 50],
    'B': [11, 21, 31, 41, 51],
    'C': [12, 22, 32, 42, 52],
    'D': [13, 23, 33, 43, 53]
}
df = pd.DataFrame(data, index=['v', 'w', 'x', 'y', 'z'])

df
```

```python
df.iloc[[1, 3],[0, 3]]
```

```python
df.iloc[1:3, 1:3]
```

```python
df
```

```python
df.loc[['w', 'y'],['B', 'D']]
```

```python
df.loc['w':'y', 'A':'C']
```

```python
df.iloc[0:3]
```
