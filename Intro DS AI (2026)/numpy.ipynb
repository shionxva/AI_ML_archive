```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
from matplotlib.ticker import LinearLocator

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       )

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)


plt.show()
```

```python
# https://lectures.scientific-python.org/intro/numpy/exercises.html#array-manipulations
# 1.3.5.4. Crude integral approximations

def f(a, b, c):
    return a ** b - c

a = np.ogrid[0:1:100j]
b = np.ogrid[0:1:100j]
c = np.ogrid[0:1:100j]

a, b, c = np.meshgrid(a, b, c)
a.shape, b.shape, c.shape

out = f(a, b, c)

out.shape
# calculate integral -> sum / number of samples
out.sum() / out.size
```

### Broadcasting

```python
M = np.ones((3, 3))
M # shape (3, 3)
```

```python
a = np.array([0, 1, 2])
a # shape (3)
```

```python
M + a[np.newaxis, :] # (1, 3)
```

```python
M + a[:, np.newaxis] # (3, 1)
```

```python
# M + a[np.newaxis, :] -> do it by repeat array
import numpy as np
#a.repeat(3, 0).reshape(3, 3)
a = np.array([[1,2], [3,4]])

a.repeat(3, axis = 0)
a = np.array([0, 1, 2])
```

```python
# 4D vs 2D

M = np.ones((3, 3, 3, 3))
M
a = np.arange(9).reshape(3,3)

# -> (1, 1, 3, 3)
M + a[np.newaxis, np.newaxis, :, :]
# -> (1, 3, 1, 3)
M + a[np.newaxis, :, np.newaxis, :]
# -> (3, 3, 1, 1)
M + a[:, :, np.newaxis, np.newaxis]

```

### Example: make data zero mean


```python
X = np.random.random((100, 2))

# calculate means of the points -> (100, 2) -> (2)
X.mean(axis = 0)

```

```python
import matplotlib.pyplot as plt

X_zero_means = X - X.mean(axis = 0) # broadcast: (100, 2) vs (2)

plt.scatter(x = X_zero_means[:,0], y=X_zero_means[:,1])

plt.scatter(x = X[:,0], y = X[:,1])
```

### Plot 2D function

```python
x = np.linspace(0, 5, 50)[np.newaxis, :]
y = np.linspace(0, 5, 50)[:, np.newaxis]

x.shape, y.shape

z = np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x.shape, y.shape, z.shape

```

```python
import matplotlib.pyplot as plt

plt.imshow(z, origin='lower', extent=[0, 5, 0, 5],
           cmap='viridis')
plt.colorbar();
```

```python
x.shape, y.shape
```

```python
# plot in 3d visualization
from matplotlib import cm

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})


x, y = np.meshgrid(x, y)

# Plot the surface.
surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm,
                       )

# Customize the z axis.
# ax.set_zlim(-1.01, 1.01)
# ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
# ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)


plt.show()
```
